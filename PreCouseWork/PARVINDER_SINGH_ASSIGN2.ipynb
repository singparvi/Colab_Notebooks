### 2.5 Use a `for` loop to calculate the BMI of each person.

Using the list you created above (`person_data`) to calculate the BMI for each individual.

The formula is as follows:

>$\frac{Weight(kg)}{Height(m)^2} = BMI$

Hint: You will iterate over one index; two of the other indices will remain the same i.e. you'll always want the first (height) and second (weight) element of each nested list.

*Additional hint:* Make sure you check the units of your input height measurements. The BMI formula uses meters (m) and your heights are in centimeters (cm). There are 100 cm in one m; to convert from cm to m, divide by 100.

```
# Write your BMI calculation loop here
# Inside the loop print out each value for the bmi 
bmi=[]
for x in range(8):
  bmicalc=person_data[x][1]/((person_data[x][2]/100)**3)
  bmi.append(bmicalc)
print (bmi)

```

### 2.6 STRETCH GOAL: Do it using list comprehension.

For loops are one way to interate through a list or nested lists. But list comprehensions are more Pythonic way to accomplish the same thing.

```
#  Calculate the BMI using a list comprehension

# One way (suggestion)
bmi=[]
bmicalc=person_data[0][1]/((person_data[0][2]/100)**3)
bmi.append(bmicalc)
bmicalc=person_data[1][1]/((person_data[1][2]/100)**3)
bmi.append(bmicalc)
bmicalc=person_data[2][1]/((person_data[2][2]/100)**3)
bmi.append(bmicalc)
bmicalc=person_data[3][1]/((person_data[3][2]/100)**3)
bmi.append(bmicalc)
bmicalc=person_data[4][1]/((person_data[4][2]/100)**3)
bmi.append(bmicalc)
bmicalc=person_data[5][1]/((person_data[5][2]/100)**3)
bmi.append(bmicalc)
bmicalc=person_data[6][1]/((person_data[6][2]/100)**3)
bmi.append(bmicalc)
bmicalc=person_data[7][1]/((person_data[7][2]/100)**3)
bmi.append(bmicalc)
print (bmi)

# More Pythonic way (suggestion)
bmi=[]
for x in range(8):
  bmicalc=person_data[x][1]/((person_data[x][2]/100)**3)
  bmi.append(bmicalc)
print (bmi)
```
