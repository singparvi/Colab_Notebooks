I can see that there are two heavy outliers in the data, but they are actually not errors. They are Greece and Spain, and after a quick google search I can see that they actually are some of the countries that are known for high unemployment rates. While I know that it would improve my statistical tests if I removed these two datapoints, I know that there is no justifiable reason to do so. 

Null hypothesis = there is no relationship between unemployment rates and mental illness
altnerative hypothesis = there is a relationship between unemployment rates and mental illness 

```python
#lets take a look at the relationship between Unemployment in 2016 and overall mental illness in 2017
import matplotlib.pyplot as plt

fig, ax = plt.subplots() 

#Plot the scatterplot#
plt.scatter(x = clear_df['Unemploy_2016 (%)'], y = clear_df['Combined_2017 (%)'])

ax.set_title('Unemployment in 2016 and mental illness in 2017')
ax.set_xlabel('Unemployment in 2016')
ax.set_ylabel('Mental illness prevalence in 2017')


plt.show()
```

So there doesn't seem to be a linear relationship between unemployment and all mental illnesses.

```python
#lets just look at the correlation coefficient
import numpy as np

r = np.corrcoef(clear_df['Unemploy_2016 (%)'], clear_df['Combined_2017 (%)'])

print(r)
```

Almost no correlation at all. 

Lets see if unemployment has a relationship with specific types of mental illness.

```python
clear_df.head()
```

```python
import seaborn as sns

#Which variables make sense to plot together in a scatterplot?
to_plot = clear_df[['GINI_year', 'HDI_2016', 'Unemploy_2016 (%)', 'Combined_2017 (%)', 'ASD_2017 (%)', 'ADHD_2017 (%)', 'DEP_2017 (%)', 'SCZ_2017 (%)','ANX_2017 (%)', 'ED_2017 (%)', 'ETOH_2017 (%)','SUICIDE_2017 (per 100.000)','BD_2017 (%)' ]]

sns.pairplot(to_plot)
```

There does not seem to be a significant relationship between unemployment and mental illness. 

So there is no significant relationship between unemployment and mental illness, as suspected. We fail to reject the null hypothesis since the p value is greater than .05.
