```
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# set the syle sheet
plt.style.use('seaborn-bright')

# create fig and ax objects
fig, ax = plt.subplots (1, 1, figsize = (8,6))

# create exponential dist.
# with 1000 samples

data_exp = stats.expon.rvs(size = 1000)

# plot the dist. and the kernel density estimate (KDE)
ax = sns.distplot(data_exp, kde = True, bins = 100)

# set the axis labels 
ax.set(xlabel= 'Exponential Dist.', ylabel= 'Frequency')


```

```
# create a list of means for each set of samples

## 
# initialize the list to hold the means 
rs_means_n25 = []
# loop 25 times, fill the list with 25 sample means
for i in range(25):
  # draw the random samples for the expo. dist.
  rs_exp = np.random.exponential(scale = 1.0, size = 1000)
  # append the mean of the random sample
  rs_means_n25.append(rs_exp.mean())

###
# initiatlize the list to hold the means 
rs_means_n100 = []

# loop 100 times, fill the list with 100 sample means
for i in range(100):
  # draw the random sample for from the exp. dist.
  rs_exp = np.random.exponential(scale = 1.0, size = 100)
  # append the mean of the random sample
  rs_means_n100.append(rs_exp.mean())


# initialize the list to hold the means
rs_means_n1000 = []

# loop 1000 times, fill the list with 1000 sample means
for i in range(1000):
  # draw random samples from the exp. dist. 
  rs_exp = np.random.exponential(scale=1.0, size=1000)
  # append the mean of the random sample
  rs_means_n1000.append(rs_exp.mean())


# create fig, ax objects
fig, (ax1, ax2, ax3) = plt.subplots(1, 3,  figsize = (10,4))

# plot each example on the correspond axis

sns.distplot(rs_means_n25, kde=True, bins=100, ax = ax1)
ax1.set(xlabel = 'n25', ylabel = 'Frequency')

sns.distplot(rs_means_n100, kde=True, bins=100, ax = ax2)
ax2.set(xlabel = 'n25', ylabel = 'Frequency')

sns.distplot(rs_means_n1000, kde=True, bins=100, ax = ax3)
ax3.set(xlabel = 'n1000', ylabel = 'Frequency')
```
