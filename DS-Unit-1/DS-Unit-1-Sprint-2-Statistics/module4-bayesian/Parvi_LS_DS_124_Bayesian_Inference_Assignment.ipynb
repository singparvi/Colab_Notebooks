### 1) In your own words, summarize the difference between Bayesian and Frequentist statistics.

The difference between Bayesian and Frequentist statistics is that in the Bayesian approach the statistics are achieved based on some prior knowledge/belief in addition to a model and through these statistics, we are only updating our belief. While in the frequentist statistics approach we are only getting statistics based on a model.

```
### YOUR CODE HERE
# find out the number of people first 
print ('The number of people on Titanic DataFrame is: ', df['Name'].count())
print ('The number of people on Survived is: ', len(df[df['Survived'] == 1]))
P_A = 342/887
print ('The Probability of Survival is  of people on Survived is: 342/887 or ', round(P_A, 2))
```

3) Calculate P(B) = P(Being a 1st class passenger)

```
### YOUR CODE HERE
print ('The number of people on Titanic DataFrame is: ', df['Name'].count())
print ('The number of people on Survived is: ', len(df[df['Pclass'] == 1]))
P_B = 216/887
print ('The Probability of Survival is  of people on Survived is: 216/887 or ', round(P_B, 2))
```

4) Calculate P(A $\cap$ B) = P(Being a 1st class passenger and surviving)

```
### YOUR CODE HERE
print (pd.crosstab(df['Survived'], df['Pclass'], margins=True))
P_A_and_B = 136/887
print ('The Probability of 1st class passenger and surviving is: 136/342 or ', round(P_A_and_B, 3))
```

5) Calculate P(A|B) = P(Surviving given that the person was a 1st class passenger)

```
### YOUR CODE HERE
# From crosstab we can see the chance of survival is 123 given 216 total of Pclass 1. Therefore:-
P_A_given_B = 136/216
print ('The Probability of Survival given person was a 1st class passenger is: or ', round(P_A_given_B, 3))
```

6) Calculate (B|A) = P(Being a first class passenger given the person survived)
