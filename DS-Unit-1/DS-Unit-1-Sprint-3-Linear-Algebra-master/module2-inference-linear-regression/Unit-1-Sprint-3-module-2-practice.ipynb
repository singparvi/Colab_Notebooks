Objective 01 - identify the appropriate hypotheses to test for a statistically significant association between two quantitative variables

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load the car crash dataset
crashes = sns.load_dataset("car_crashes")

mycols = ['total', 'speeding', 'alcohol']
crashes = crashes[mycols]
crashes.head()
```

Objective 02 - conduct an interpret a t-test for the slope parameter and make the connection between the t-test for a population mean and slope coefficient

```
import warnings
warnings.filterwarnings('ignore')
import statsmodels.api as sm

Y = crashes['total']
X = crashes['alcohol']

# create the model with X and Y data
X = sm.add_constant(X)
model = sm.OLS(Y,X)

# fit the model
results = model.fit()

#look at the results that include the t-statistics
print (results.t_test([1,0]))
```

Objective 03 - identify the appropriate parts of the output of a linear regression model and use them to build a confidence interval for the slope term 

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

#load the crash data set
crashes = sns.load_dataset('car_crashes')

my_cols = ['total', 'speeding', 'alcohol']

crashes = crashes[my_cols]
crashes.head()
```

```
import statsmodels.api as sm

Y = crashes['total']
X = crashes['alcohol']

# create the model with X and Y data
X = sm.add_constant(X)
model = sm.OLS(Y,X)

# fit the model
results = model.fit()

#look at the results that include the t-statistics
print (results.t_test([1,0])) # ??? how to read the confidence interval data from below?
```

```
# create the scatter plot with the confidence interval 

sns.lmplot(data = crashes, x = 'alcohol', y = 'total', ci = 95)
```
